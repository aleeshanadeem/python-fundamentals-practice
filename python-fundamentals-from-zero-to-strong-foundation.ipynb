🔹 Operators
</h1>

* **What:**  In Python, operators are symbols that perform operations on variables and values. 
 ➡️                             ****Operator = Action perform tool****

* **Define:** An operator is a symbol used to perform mathematical or logical operations on data.

* **How:**  Operators work between values (called operands).

* **Why Use:**

➡️ Perform calculations

➡️ Compare values

➡️ Make decisions

➡️ Control program logic




| Type                  | Operators                           | Use                          | Description |
|-----------------------|-------------------------------------|------------------------------|-------------|
| Arithmetic            | +  -  *  /  %  **  //               | Mathematical calculations    | Perform addition, subtraction, multiplication, division, remainder, power, and floor division. |
| Comparison            | >  <  ==  !=  >=  <=                | Compare values               | Compare two values and return True or False. |
| Logical               | and  or  not                        | Combine conditions           | Used to combine multiple conditions and return Boolean results. |
| Assignment            | =  +=  -=  *=  /=  %=  **=  //=     | Assign or update values      | Assign values to variables and update existing values. |
| Membership            | in  not in                          | Check existence in sequence  | Returns True if a value exists in a list, tuple, or string. |
| Identity              | is  is not                          | Check object identity        | Checks whether two variables refer to the same memory location. |

* **Let's see it now as a Code and Example**

```python
# Sample Variables
a = 10
b = 3
numbers = [1, 2, 3, 4]

# 1️⃣ Arithmetic Operators
print("Output:\n")
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Modulus:", a % b)
print("Power:", a ** b)
print("Floor Division:", a // b)
```

🔹 Conditional Statements</h1>



* **What:**    In Python, conditions are used to make **decisions** in a program.
* **Define:**  A conditional statement is used to **execute** a block of code only if a **specific condition** is true.
* **How:**
Conditions use keywords like:

   **1. If**  → First decision

   **2. elif**  → Second decision

  **3. else**  → Final backup plan 
* **Structure**
  
  

* **Use**

**1.** Make decisions

**2.** Control program flow

**3.** Execute code selectively

**4.** Build intelligent systems
* **Example:**


```python
marks = 75

if marks >= 50:
    print("Pass😁")
else:
    print("Fail😟")

print("Beacuse condition is true if condition is falus to fail code run.")
```

```python
hunger = 8

if hunger > 8:
    print("Give extra large pizza 🍕🍕")
elif hunger > 5:
    print("Give medium pizza 🍕")
else:
    print("Give juice only 🧃")
```

🔹 Loops Statements</h1>



* **What:**   In Python, loops are used to **repeat** a block of code multiple times.
  
* **Define:**  A loop is a **control structure** that executes a block of **code repeatedly** until a condition is satisfied.

* **Work:**
              If the condition is True, run the code; otherwise, the code stops  
* **Types:**

| Type of Loop | Structure | Description | When to Use |
|--------------|------------|-------------|--------------|
| for loop     | for variable in sequence: | Repeats code for a fixed number of times or over a sequence (like list, range, string). | When number of repetitions is known. |
| while loop   | while condition: | Repeats code as long as the condition is True. | When repetition depends on a condition. |

* **Use**

***1.*** Repeat tasks automatically

***2.*** Save time and code

***3.*** Process large data

***4.*** Work with lists and datasets

***5.*** Build efficient programs

* **Example**

**1. For Loop**

```python
print("For Loop\n")
for i in range(5):
    print("Hello")
```

**2. While Loop**

Run the code as long as condition is true.

```python
print("OUTPUT")
x = 1
while x <= 5:
    print(x)
    x += 1
```

🔹 Tuple
</h1>



* **What:**   In Python, a tuple is a data structure used to store multiple values in a single variable — just **like a list**, but with one **major difference.**

* **Define:**
A tuple is an ordered, **immutable** collection of items enclosed in **parentheses ().**

* **Why:**

 We use tuples to:

➡️ Store fixed data

➡️ Protect data from modification

➡️ Improve performance (faster than lists)

➡️ Represent constant values

➡️ Return multiple values from functions

* **How:**
  The hot print() is created.

* **Example:**

```python
numbers = (10, 20, 30)
print("OUTPUT:")
print(numbers)
```

| Feature        | List                         | Tuple                        |
|---------------|------------------------------|------------------------------|
| Brackets      | []                           | ()                           |
| Mutable       | Yes (Changeable)             | No (Immutable / Fixed)       |
| Performance   | Slightly Slower              | Slightly Faster              |
| Memory Usage  | Uses More Memory             | Uses Less Memory             |
| Use Case      | Dynamic / Changeable Data    | Fixed / Constant Data        |
| Example       | [1, 2, 3]                    | (1, 2, 3)                    |
